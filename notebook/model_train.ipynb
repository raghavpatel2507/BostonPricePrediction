```python
# independent and dependent fetures
X=df.drop('price',axis=1)
y=df['price']
```

```python
X.head()
```

```python
y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
linear_regeression=LinearRegression()
linear_regeression.fit(X_train_scaled,y_train)
```

```python
linear_regeression.coef_
```

```python
linear_regeression.intercept_
```

```python
print(linear_regeression.get_params())
```

```python
pred1=linear_regeression.predict(X_test_scaled)
```

```python
pred1
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,pred1))
print(mean_squared_error(y_test,pred1))
print(r2_score(y_test,pred1))
```

```python
# random forest
reg2=RandomForestRegressor()
reg2.fit(X_train_scaled,y_train)
```

```python
pred2=reg2.predict(X_test_scaled)
```

```python
pred2
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,pred2))
print(mean_squared_error(y_test,pred2))
print(r2_score(y_test,pred2))
```

```python
sns.regplot(x=y_test,y=pred2,data=df)
```

```python
reg3=DecisionTreeRegressor()
reg3.fit(X_train_scaled,y_train)
```

```python
pred3=reg3.predict(X_test_scaled)
```

```python
pred3
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,pred3))
print(mean_squared_error(y_test,pred3))
print(r2_score(y_test,pred3))
```

```python
sns.regplot(x=y_test,y=pred3,data=df)
```

```python
reg4=SVR()
reg4.fit(X_train_scaled,y_train)
```

```python
pred4=reg4.predict(X_test_scaled)
```

```python
pred4
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
print(mean_absolute_error(y_test,pred4))
print(mean_squared_error(y_test,pred4))
print(r2_score(y_test,pred4))
```

```python
sns.regplot(x=y_test,y=pred4,data=df)
```
